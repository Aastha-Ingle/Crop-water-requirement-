```python
# Reading the dataset
import pandas as pd
df = pd.read_csv('crop_water_requirement_updated.csv')
display (df)
```

```python
df.rename (columns = {"Water_Requirement (liters/hectare/day)" : "Water_Required" }, inplace = True)
```

```python
df_cleaned = df.dropna (subset = ["Water_Required"])
display (df_cleaned)
```

```python
# Handling duplicates
# count the duplicate rows
print (df_cleaned.duplicated().sum())
```

```python
# drop the duplicate rows keeping the first occurence
df_cleaned = df_cleaned.drop_duplicates()

# resetting the index after dropping
df_cleaned = df_cleaned.reset_index (drop = True)

display (df_cleaned)
```

```python
y = df_cleaned["Water_Required"]
x = df_cleaned.drop("Water_Required", axis = 1)
display (x)
```

```python
# Detecting outliers
numeric_column = x.select_dtypes(include=['number']).columns  # separating numeric columns

for col in numeric_column:         # traversing through the columns
  Q1 = x[col].quantile(0.25)       # calculating fist quartile and
  Q3 = x[col].quantile(0.75)       # third quartile of each column

  IQR = Q3 - Q1                    # Interquartile range

  lower = Q1 - 1.5*IQR           
  upper = Q3 + 1.5*IQR           

  for i in range(len(x)):          # traversing through the rows of each column
    val = x.iloc[i][col]           # store the value of each cell of a row in 'val'
    if val < lower or val > upper:                  # check if 'val' lie within the lower bound and upper bound range
      x.iloc[i, x.columns.get_loc(col)] = pd.NA     # if not then change the dataframe to NA


display (x)
```

```python
x.reset_index(drop=True, inplace=True)
```

```python
print(x.columns)
```

```python
# Identify categorical columns
categorical_cols = ["District", "Crop_Type", "Soil_Type"]

# Convert categorical columns to one-hot encoded columns
x = pd.get_dummies(x, columns=categorical_cols)
```

```python
# filling the cells with mean of their columns
x = x.fillna(x.mean())
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
from sklearn.ensemble import VotingRegressor, RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# Initialize individual models
rf = RandomForestRegressor(random_state=42)
dt = DecisionTreeRegressor(random_state=42)
xgb = XGBRegressor(random_state=42, verbosity=0)

# Train each model individually
rf.fit(x_train, y_train)
dt.fit(x_train, y_train)
xgb.fit(x_train, y_train)

# Predict
rf_pred = rf.predict(x_test)
dt_pred = dt.predict(x_test)
xgb_pred = xgb.predict(x_test)

# Evaluate RMSE
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))
dt_rmse = np.sqrt(mean_squared_error(y_test, dt_pred))
xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_pred))

# Print individual RMSEs
print(f"Random Forest RMSE: {rf_rmse:.2f}")
print(f"Decision Tree RMSE: {dt_rmse:.2f}")
print(f"XGBoost RMSE: {xgb_rmse:.2f}")

# Voting Regressor
voting_model = VotingRegressor([
    ('rf', rf),
    ('dt', dt),
    ('xgb', xgb)
])

voting_model.fit(x_train, y_train)
voting_pred = voting_model.predict(x_test)
voting_rmse = np.sqrt(mean_squared_error(y_test, voting_pred))
print(f"Voting Regressor RMSE: {voting_rmse:.2f}")

# Best model based on RMSE
rmse_dict = {
    'Random Forest': rf_rmse,
    'Decision Tree': dt_rmse,
    'XGBoost': xgb_rmse,
    'Voting Regressor': voting_rmse
}
best_model = min(rmse_dict, key=rmse_dict.get)
print(f"\n✅ Best Model: {best_model} with RMSE = {rmse_dict[best_model]:.2f}")
```
